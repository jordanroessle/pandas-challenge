```python
# Dependencies and Setup
import pandas as pd
import os

# File to Load (Remember to Change These)
file_to_load = os.path.join("Resources", "purchase_data.csv")

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
#unique in case multiple purchases by same player
total_players = len(purchase_data.SN.unique())

#outputing like this to match what was provided
pd.DataFrame({"Total Players":total_players}, index = [0])
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#calculate desired values
unique_items = len(purchase_data["Item ID"].unique())
average_price = round(purchase_data.Price.mean(),2)
total_purchases = purchase_data.shape[0]
total_revenue = purchase_data.Price.sum()

#insert calculated values into a dataframe
analysis_total = pd.DataFrame({"unique_items":unique_items, 
                               "average_price":average_price,
                              "total_purchases": total_purchases, 
                              "total_revenue":total_revenue}, index=[0])
#formatting
analysis_total.total_revenue = analysis_total.total_revenue.map("${:.2f}".format)
analysis_total.average_price = analysis_total.average_price.map("${:.2f}".format)

#output
analysis_total
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#drop duplicates to get rid of multiple purchases from same player
gender_df = purchase_data[["SN", "Gender"]].drop_duplicates()

#intitalize data frame
gender_demo = pd.DataFrame()

#Calculate count and percentage of players
gender_demo["Total_Count"] = gender_df.Gender.value_counts()
gender_demo["Percentage_of_Players"] = 100 * gender_demo.Total_Count / total_players

#Formatting
gender_demo["Percentage_of_Players"] = gender_demo["Percentage_of_Players"].map("{:.2f}%".format)

#Output
gender_demo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#initialize groupby and dataframe
gender_groupby = purchase_data.groupby("Gender")
gender_analysis = pd.DataFrame(index=gender_groupby.SN.count().index)

#Calcualte desired values, while adding column to the data frame
gender_analysis["Purchase_Count"] = gender_groupby["SN"].count() 
gender_analysis["Average_Purchase_Price"] = gender_groupby.Price.mean()
gender_analysis["Total_Purchase_Value"] = gender_groupby.Price.sum()
gender_analysis["Avg_Total_Purchase_per_Person"] = gender_analysis.Total_Purchase_Value / gender_demo.Total_Count

#formatting
gender_analysis["Average_Purchase_Price"] = gender_analysis["Average_Purchase_Price"].map("${:.2f}".format)
gender_analysis["Total_Purchase_Value"] = gender_analysis["Total_Purchase_Value"].map("${:.2f}".format)
gender_analysis["Avg_Total_Purchase_per_Person"] = gender_analysis["Avg_Total_Purchase_per_Person"].map("${:.2f}".format)

#output
gender_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#max of data set is 45, min is 7, set up bins to capture both
bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]

#matching provided labels
labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

#grabbing unique users by age
unique_id_age = purchase_data[["Age", "SN"]].drop_duplicates()

#categorize using pd.cut
unique_id_age["Age_Range"] = pd.cut(unique_id_age.Age, bins=bins, labels=labels)

#intialize output dataframe and age groupby
age_demo = pd.DataFrame(index=labels)
age_groupby = unique_id_age.groupby("Age_Range")

#calculate desired values
age_demo["Total_Count"] = age_groupby.Age.count()
age_demo["Percentage_of_Players"] = 100 * age_demo.Total_Count / total_players

#formatting
age_demo["Percentage_of_Players"] = age_demo["Percentage_of_Players"].map("{:.2f}%".format)

#output
age_demo
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#have to cut again, this time with all purchases
purchase_data["Age_Ranges"] = pd.cut(purchase_data.Age, bins=bins, labels=labels)

#initialize output dataframe and groupby
age_groupby = purchase_data.groupby("Age_Ranges")
age_analysis = pd.DataFrame(index=age_groupby.Age.count().index)

#Calculate desired values
age_analysis["Purchase_Count"] = age_groupby.Age.count() 
age_analysis["Average_Purchase_Price"] = age_groupby.Price.mean()
age_analysis["Total_Purchase_Value"] = age_groupby.Price.sum()
age_analysis["Avg_Total_Purchase_per_Person"] = age_analysis["Total_Purchase_Value"] / age_demo["Total_Count"]

#formatting
age_analysis["Average_Purchase_Price"] = age_analysis["Average_Purchase_Price"].map("${:.2f}".format)
age_analysis["Total_Purchase_Value"] = age_analysis["Total_Purchase_Value"].map("${:.2f}".format)
age_analysis["Avg_Total_Purchase_per_Person"] = age_analysis["Avg_Total_Purchase_per_Person"].map("${:.2f}".format)

age_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#incase you want to change how many top spenders to look at later
top_number_spenders = 5

#initialize output dataframe and groupby
SN_groupby = purchase_data.groupby("SN")
#grabs the top 5 spenders and puts them as an index
top_spenders = pd.DataFrame(index = SN_groupby.Price.sum().sort_values(ascending=False).head(top_number_spenders).index)

#calculate desired values
top_spenders["Purchase_Count"] = SN_groupby.Price.count()
top_spenders["Average_Purchase_Price"] =  SN_groupby.Price.mean()
top_spenders["Total_Purchase_Value"] =  SN_groupby.Price.sum()

#formatting
top_spenders["Average_Purchase_Price"] = top_spenders["Average_Purchase_Price"].map("${:.2f}".format)
top_spenders["Total_Purchase_Value"] = top_spenders["Total_Purchase_Value"].map("${:.2f}".format)

#output
top_spenders
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#incase you want to change how many most popular items to look at later
number_popular_items = 5 

#initialize output dataframe and groupby
item_id_groupby = purchase_data.groupby(["Item ID", "Item Name"])
#4 items have 9 purchases, so my data frame shows one different than the example provided
popular_items = pd.DataFrame(index=item_id_groupby["Item Name"].count().sort_values(ascending=False).head(number_popular_items).index)

#calculate desired values
popular_items["Purchase_Count"] = item_id_groupby.Price.count()
popular_items["Item_Price"] = item_id_groupby.Price.mean()
popular_items["Total_Purchase_Value"] = item_id_groupby.Price.sum()

#formatting
popular_items["Item_Price"] = popular_items["Item_Price"].map("${:.2f}".format)
popular_items["Total_Purchase_Value"] = popular_items["Total_Purchase_Value"].map("${:.2f}".format)

#output
popular_items
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#incase you want to change how many most popular items to look at later
number_profitable_items = 5 

#initialize output dataframe and groupby
profitable_groupby = purchase_data.groupby(["Item ID", "Item Name"])
profitable_items = pd.DataFrame(index=profitable_groupby.Price.sum().sort_values(ascending=False).head(number_profitable_items).index)

#Calculate desired values
profitable_items["Purchase_Count"] = profitable_groupby.Price.count()
profitable_items["Item_Price"] = profitable_groupby.Price.mean()
profitable_items["Total_Purchase_Value"] = profitable_groupby.Price.sum()

#formatting
profitable_items["Item_Price"] = profitable_items["Item_Price"].map("${:.2f}".format)
profitable_items["Total_Purchase_Value"] = profitable_items["Total_Purchase_Value"].map("${:.2f}".format)

#output
profitable_items
```
